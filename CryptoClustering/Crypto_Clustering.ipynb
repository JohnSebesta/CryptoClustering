```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id"
)
 
# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
 
```

```python
# Check the column names in the DataFrame
df_market_data.columns
 
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# This standardizes features by removing the mean and scaling to unit variance
crypto_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Use the same column names as the original DataFrame
crypto_scaled_df = pd.DataFrame(
    crypto_scaled,
    columns=df_market_data.columns
)
 
# Copy the crypto names from the original data
c_names = df_market_data.index
 
# Set the coin_id column as index
crypto_scaled_df['coin_id'] = c_names
crypto_scaled_df = crypto_scaled_df.set_index('coin_id')
 
# Display sample data
crypto_scaled_df.head()
 
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k values from 1 to 11
k = list(range(1, 12))
 
```

```python
# Create an empty list to store the inertia values
inertia = []
 
# Create a for loop to compute the inertia with each possible value of k
# Inertia is the sum of squared distances of samples to their closest cluster center
for i in k:
    # Initialize the K-Means model with the current value of k
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
 
    # Fit the model to the scaled data
    k_model.fit(crypto_scaled_df)
 
    # Append the inertia to the list
    inertia.append(k_model.inertia_)
 
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict = {'k': k, 'inertia': inertia}
 
# Create a DataFrame with the data for plotting
elbow_df = pd.DataFrame(elbow_dict)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k
elbow = elbow_df.hvplot.line(
    x='k',
    y='inertia',
    xticks=k,
    title='Elbow Curve for K-Means Clustering (Original Scaled Data)'
)
 
# Display the plot
elbow
 
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** From the elbow curve, the best value for k appears to be 4, as this is where 
# the rate of decrease in inertia slows down significantly, forming an "elbow" in the plot.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=1, n_init=10)
```

```python
# Fit the K-Means model using the scaled data
k_model.fit(crypto_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
clusters = k_model.predict(crypto_scaled_df)
 
# Print the resulting array of cluster values
print(clusters)
 
```

```python
# Create a copy of the DataFrame
crypto_scaled_df_copy = crypto_scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
crypto_scaled_df_copy['cluster'] = clusters
 
# Display sample data
crypto_scaled_df_copy.head()
 
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
crypto_scatter = crypto_scaled_df_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="cluster",
    hover_cols=["coin_id"],
    title="Cryptocurrency Clusters - Original Data"
)
 
# Display the scatter plot
crypto_scatter
 
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`
# PCA reduces dimensionality while preserving as much variance as possible
pca = PCA(n_components=3)
 
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components
crypto_pca = pca.fit_transform(crypto_scaled_df)
 
# View the first five rows of the PCA data
print(crypto_pca[:5])
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is approximately 
# 88.44% of the original data's variance, meaning our dimensionality reduction preserves 
# most of the information from the original seven features.

```python
# Calculate the total explained variance
total_explained_variance = sum(explained_variance)
print(f"Total explained variance: {total_explained_variance:.4f} or {total_explained_variance*100:.2f}%")
```

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(
    crypto_pca, 
    columns=['PC1', 'PC2', 'PC3']
)
 
# Copy the crypto names from the original data
pca_df['coin_id'] = df_market_data.index
 
# Set the coin_id column as index
pca_df = pca_df.set_index('coin_id')
 
# Display sample data
pca_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []
 
# Create a for loop to compute the inertia with each possible value of k
for i in k:
    # Initialize the K-Means model with the current value of k
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
 
    # Fit the model to the PCA data
    k_model.fit(pca_df)
 
    # Append the inertia to the list
    inertia.append(k_model.inertia_)
 
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict = {'k': k, 'inertia': inertia}
 
# Create a DataFrame with the data for plotting
elbow_df = pd.DataFrame(elbow_dict)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k
elbow_pca = elbow_df.hvplot.line(
    x='k',
    y='inertia',
    xticks=k,
    title='Elbow Curve for K-Means Clustering (PCA Data)'
)
 
# Display the plot
elbow_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** From the elbow curve of the PCA data, the best value for k still appears to be 4,


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:**  No, the optimal k value of 4 is the same for both the original scaled data and the PCA data.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=1, n_init=10)
```

```python
# Fit the K-Means model using the PCA data
k_model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_clusters = k_model.predict(pca_df)

# Print the resulting array of cluster values.
print(pca_clusters)
```

```python
# Create a copy of the scaled PCA DataFrame
pca_df_copy = pca_df.copy()
 
# Add a new column to the DataFrame with the predicted clusters
pca_df_copy['cluster'] = pca_clusters
 
# Display sample data
pca_df_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
crypto_scatter_pca = pca_df_copy.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="cluster",
    hover_cols=["coin_id"],
    title="Cryptocurrency Clusters - PCA Data"
)
 
# Display the scatter plot
crypto_scatter_pca
 
```
